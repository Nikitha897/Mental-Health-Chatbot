```python
import faiss
import numpy as np
import wikipediaapi
from sentence_transformers import SentenceTransformer
from ctransformers import AutoModelForCausalLM
wiki_wiki = wikipediaapi.Wikipedia(language="en", user_agent="MentalHealthChatbot/1.0")
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")  # Lightweight & fast embedding model
class WikipediaRetriever:
    def __init__(self):
        self.knowledge_base = []
        self.index = faiss.IndexFlatL2(384)  # 384 dimensions (from MiniLM)
        self.embeddings = []

    def add_to_knowledge_base(self, topic):
        """Fetch a Wikipedia summary, convert to embeddings, and store in FAISS"""
        page = wiki_wiki.page(topic)
        if not page.exists():
            return "No relevant data found."

        summary = page.summary[:500]  # Keep it concise
        embedding = embedding_model.encode([summary])  # Convert text to vector
        self.index.add(np.array(embedding, dtype=np.float32))
        self.knowledge_base.append(summary)
        self.embeddings.append(embedding)
        return summary

    def retrieve_relevant_info(self, user_query):
        """Find the most relevant Wikipedia summary for the user's question"""
        query_embedding = embedding_model.encode([user_query])
        D, I = self.index.search(np.array(query_embedding, dtype=np.float32), 1)  # Find closest match
        return self.knowledge_base[I[0][0]] if len(self.knowledge_base) > 0 else "No relevant data found."
```

```python
def mental_health_chatbot():
    retriever = WikipediaRetriever()

    # Preload some common mental health topics
    retriever.add_to_knowledge_base("Cognitive Behavioral Therapy")
    retriever.add_to_knowledge_base("Mindfulness")
    retriever.add_to_knowledge_base("Anxiety")
    retriever.add_to_knowledge_base("Depression")

    print("🧠 Mental Health Chatbot is active. Type 'exit' to quit.")

    while True:
        user_input = input("\nYou: ")
        if user_input.lower() == "exit":
            print("\nChatbot: Take care! I'm always here if you need to talk. ❤️")
            break

        # Retrieve relevant Wikipedia info from FAISS
        context = retriever.retrieve_relevant_info(user_input)

        # Generate AI response with context
        prompt = f"""
        Using this reference: {context}

        You are a mental health assistant. A user asked:
        "{user_input}"
        Provide a supportive and informative response:
        """

        # Generate response
        response = model(prompt)
        print("\nChatbot:", response.strip())

# Run the chatbot
mental_health_chatbot()
```
